```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

## Image Resizing and Normalization

```python
path = "C:/Users/Katon/Documents/JHU/CreatingAIEnabledSystems/Assignment2/BikePhotos/"

new_size = (425,425)

fig = plt.figure()


for i in range(1,5):
    # handle both jpg and png
    img = cv2.imread(path + "BikeShot" + str(i) + ".jpg")
    if img is None:
        img = cv2.imread(path + "BikeShot" + str(i) + ".png")
    
    height = img.shape[0]
    width = img.shape[1]
    channels = img.shape[2]
    
    print("Image" + str(i) +":")
    print(img.shape)
    
    
    # Image Resizing
    src = img
    resized_img = cv2.resize(src, new_size, interpolation = cv2.INTER_AREA)
    
    print(resized_img.shape)
    plt.imshow(resized_img)
    
    # Normalize image
    # put all pixels between 0 and 1
    #norm_img = resized_img / 250
    # subtract mean
    norm_img = resized_img - np.mean(resized_img)
    # divide by standard deviation
    norm_img = norm_img / np.std(norm_img)
    
    # print normalized images
    plt.imshow(norm_img)
    
    
    plt.show() 
```
